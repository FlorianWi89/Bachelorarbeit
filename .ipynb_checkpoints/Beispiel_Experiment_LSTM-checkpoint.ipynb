## Util Functions for Sensor Fault Generation

```python

#Type 1: constant positive / negative offset to true sensor measurement
def create_type_one_fault(data):
    #check the lenght of the sequence
    l = len(data)
    
    offset = np.random.randint(-5,5)
    #choose random timeframe for fault
    lower_idx = np.random.randint(l/2, l-1)
    upper_idx = lower_idx + np.random.randint(1000,2500)
    
    #implement fault
    data[lower_idx:upper_idx] += offset
    return data, lower_idx, upper_idx



#Type 2: gaussian noise
def create_type_two_fault(data):
    #check the lenght of the sequence
    l = len(data)
    
    #choose random timeframe for fault
    lower_idx = np.random.randint(l/2, l-1)
    upper_idx = lower_idx + np.random.randint(1000,2500)
    
    noise = np.random.normal(0, 1, (upper_idx - lower_idx))

    #implement fault
    data[lower_idx:upper_idx] += noise
    return data, lower_idx, upper_idx



#Type 3: sensor power failures, results in 0 values
def create_type_three_fault(data):
    #check the lenght of the sequence
    l = len(data)
    
    
    #choose random timeframe for fault
    lower_idx = np.random.randint(l/2, l-1)
    upper_idx = lower_idx + np.random.randint(1000,2500)
    
    #implement fault
    data[lower_idx:upper_idx] = 0 
    return data, lower_idx, upper_idx



#Type 4: linear proportional offset to true measurement
def create_type_four_fault(data):
    #check the lenght of the sequence
    l = len(data)
    
    
    #choose random timeframe for fault
    lower_idx = np.random.randint(l/2, l-1)
    upper_idx = lower_idx + np.random.randint(1000,2500)
    
    #implement fault
    data[lower_idx:upper_idx] *= 1.03
    return data, lower_idx, upper_idx



#Type 5: Sensor drift, increasing measurement value until a maximum is reached
def create_type_five_fault(data):
    #check the lenght of the sequence
    l = len(data)
    
    
    #choose random timeframe for fault
    lower_idx = np.random.randint(l/2, l-1)
    upper_idx = lower_idx + np.random.randint(1000,2500)
    
    #implement fault
    offset = 0.002
    
    for i in range(lower_idx, upper_idx):
        data[i] += offset
        offset += 0.003
    return data, lower_idx, upper_idx




##########################

##########################


#function to get Model data from Scenario

def get_model_data(s:Scenario, fault_type=1, target_sensor=1):
    #step 1: prepare X, y_true (may contain a sensor fault)
    X = []
    y = []
    low_idx = 0
    high_idx = 0
    
    target_sensor = "Node_" + str(target_sensor)
    
    #do we have to implement a sensor fault ?
    if fault_type == 0:
        y = s.pressures[target_sensor] #no fault
    
    elif fault_type == 1:
        y, low_idx, high_idx = create_type_one_fault(s.pressures[target_sensor])
        
    elif fault_type == 2:
        y, low_idx, high_idx = create_type_two_fault(s.pressures[target_sensor])
        
    elif fault_type == 3:
        y, low_idx, high_idx = create_type_three_fault(s.pressures[target_sensor])
        
    elif fault_type == 4:
        y, low_idx, high_idx = create_type_four_fault(s.pressures[target_sensor])
        
    elif fault_type == 5:
        y, low_idx, high_idx = create_type_five_fault(s.pressures[target_sensor])
        
    s.pressures = s.pressures.drop([target_sensor],axis=1)
    
    for i in range(len(s.pressures)):
        x = s.pressures.loc[i].to_numpy()
        X.append(x)
    
    return X, y,low_idx, high_idx



def get_train_test_split(X,y, train_size=5000):
    train_size = 6000
    
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    return X_train, X_test, y_train, y_test
```

## Create Train-Test Split for Model Training

```python
X_train, X_test, y_train, y_test = get_train_test_split(X,y)

X_train = np.array(X_train)
X_test = np.array(X_test)

y_train = np.array(y_train)
y_test = np.array(y_test)

print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

```python
#for our model we have to reshape the data

X_train = X_train.reshape((6000,31,1))
X_test = X_test.reshape((11520,31,1))

y_train = y_train.reshape((6000,1))
y_test = y_test.reshape((11520,1))

print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

## Create and train the LSTM Model

```python
LSTM_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(31, return_sequences=False),
    tf.keras.layers.Dense(units=1)
])

LSTM_model.compile(loss = tf.keras.losses.MeanSquaredError(),
                   optimizer = tf.keras.optimizers.Adam(), 
                   metrics=[tf.keras.metrics.mean_absolute_error])

```

```python
history = LSTM_model.fit(X_train, y_train, validation_split=0.2, epochs=15, batch_size=16)
```

## Inspect the models performance

```python
plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])
plt.title('model performance')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

## Let the Model predict the virtual sensor data

```python
prediction = LSTM_model.predict(X_test).flatten()
```

```python
pd.DataFrame(prediction).plot(figsize=(20,5))
```
