```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.linear_model import Ridge, LinearRegression
```

```python
TRAIN_DATA_PATH = 'train_data_sensor_fault.parquet.gzip'
TARGET_ID = 0
TEST_DATA_PATH = '/Users/florianwicher/Desktop/TestData/SensorFaultScenariosTest_0'
train_percentage = 0.15
batch_size = 128
epochs = 1
lr = 0.001
```

```python
#load train data
train_data = pd.read_parquet(TRAIN_DATA_PATH).to_numpy().astype(np.float64)
features = list(range(train_data.shape[1]));features.remove(TARGET_ID)

train_data = train_data[: int(train_percentage * len(train_data))]

X_train, y_train = train_data[:,features], train_data[:,TARGET_ID]

Ridge = Ridge().fit(X_train, y_train)
LinReg = LinearRegression().fit(X_train, y_train)

X_train = np.array(X_train).reshape(X_train.shape[0], X_train.shape[1], 1)
y_train = np.array(y_train).reshape(y_train.shape[0],1)

print(X_train.shape)
print(y_train.shape)
```

```python
RNN = tf.keras.Sequential()
RNN.add(tf.keras.layers.SimpleRNN(30, return_sequences=True))
RNN.add(tf.keras.layers.Dropout(0.2))
RNN.add(tf.keras.layers.SimpleRNN(30, return_sequences=False))
RNN.add(tf.keras.layers.Dense(20, activation='sigmoid'))
RNN.add(tf.keras.layers.Dense(1))

RNN.compile(loss = tf.keras.losses.MeanAbsoluteError(),
                   optimizer =tf.keras.optimizers.legacy.Adam(learning_rate=lr),
                   metrics=['mean_squared_error'])

with tf.device('/cpu:0'):
    RNN.fit(X_train,y_train, batch_size=batch_size, epochs=epochs)
    
RNN.summary()
```

```python
GRU = tf.keras.Sequential()
GRU.add(tf.keras.layers.GRU(30, return_sequences=True))
GRU.add(tf.keras.layers.Dropout(0.2))
GRU.add(tf.keras.layers.GRU(30, return_sequences=False))
GRU.add(tf.keras.layers.Dense(20, activation='sigmoid'))
GRU.add(tf.keras.layers.Dense(1))

GRU.compile(loss = tf.keras.losses.MeanAbsoluteError(),
                   optimizer =tf.keras.optimizers.Adam(learning_rate=lr),
                   metrics=['mean_squared_error'])

with tf.device('/cpu:0'):
    GRU.fit(X_train,y_train, batch_size=batch_size, epochs=epochs)
    
GRU.summary()
```

```python
LSTM = tf.keras.Sequential()
LSTM.add(tf.keras.layers.LSTM(30, return_sequences=True))
LSTM.add(tf.keras.layers.Dropout(0.2))
LSTM.add(tf.keras.layers.LSTM(30, return_sequences=False))
LSTM.add(tf.keras.layers.Dense(20, activation='sigmoid'))
LSTM.add(tf.keras.layers.Dense(1))

LSTM.compile(loss = tf.keras.losses.MeanAbsoluteError(),
                   optimizer =tf.keras.optimizers.Adam(learning_rate=lr),
                   metrics=['mean_squared_error'])

with tf.device('/cpu:0'):
    LSTM.fit(X_train,y_train, batch_size=batch_size, epochs=epochs)
    
LSTM.summary()
```

```python
constant = 1
drift = 2
normal = 3
percentage = 4
stuckzero = 5

start = 8200
end = 9200
steps = 1
path = '/Users/florianwicher/Desktop/resultData/Sensor_fault'
```

```python
test_data_path = os.path.join(TEST_DATA_PATH, f'{constant}_sensor_fault_test_data.parquet.gzip')
test_data = pd.read_parquet(test_data_path).to_numpy().astype(np.float64)

X_test, y_test = test_data[:,features], test_data[:,TARGET_ID]

ridge_pred = Ridge.predict(X_test)
linreg_pred = LinReg.predict(X_test)


X_test = np.array(X_test).reshape(X_test.shape[0], X_test.shape[1], 1)

with tf.device('/cpu:0'):
    rnn_pred = RNN.predict(X_test, batch_size=512).flatten()
    
with tf.device('/cpu:0'):
    gru_pred = GRU.predict(X_test, batch_size=512).flatten()
    
with tf.device('/cpu:0'):
    lstm_pred = LSTM.predict(X_test, batch_size=512).flatten()
    
plt.figure(figsize=(15, 3))
plt.plot(y_test[start:end:steps], c='red', label='y true');
plt.plot(rnn_pred[start:end:steps], c='orange', label='RNN prediction');
plt.plot(gru_pred[start:end:steps], c='green', label='GRU prediction');
plt.plot(lstm_pred[start:end:steps], c='blue', label='LSTM prediction');
plt.plot(ridge_pred[start:end:steps], c='c', label='Ridge prediction');
plt.plot(linreg_pred[start:end:steps], c='m', label='LinReg prediction');

plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Pressure')
plt.legend(loc='upper left')
#plt.savefig(os.path.join(path, 'constant_error_predictions.pdf'), bbox_inches="tight")
plt.show();
```
