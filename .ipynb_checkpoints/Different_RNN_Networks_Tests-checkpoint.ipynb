## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from LeakDbScenario import Scenario, ScenarioWithPreprocessing
from LeakDbScenario import get_scenarios_with_without_leakages
from utils import create_type_one_fault, create_type_two_fault, create_type_three_fault
from utils import create_type_four_fault, create_type_five_fault, get_model_data, get_train_test_split
from utils import get_possible_fault_points
```

## Load and prepare Scenario

```python
scenario = Scenario('10','/Users/florianwicher/Desktop/LeakDB/Hanoi_CMH/')

X, y, l_idx, h_idx = get_model_data(scenario, fault_type=4, target_sensor=17)
```

```python
sensor_data_with_fault = pd.DataFrame({'Sensor_Data' : y})
sensor_data_with_fault.plot(figsize=(20,5))
```

```python
# prep data for training

RNN_UNITS = 31
TRAIN_SIZE = 7000

X_train, X_test, y_train, y_test = get_train_test_split(X, y, train_size = TRAIN_SIZE, network_units= RNN_UNITS)
```

```python
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

## Train + Predict without Scaling

```python
LSTM_model = tf.keras.models.Sequential([
    tf.keras.layers.GRU(RNN_UNITS, return_sequences=True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.GRU(RNN_UNITS, return_sequences=False),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units=31),
    tf.keras.layers.Dense(1)
])

LSTM_model.compile(loss = 'mean_absolute_error',
                   optimizer = 'adam', 
                   metrics=['mean_absolute_error'])
```

```python
# fitting the RNN
with tf.device('/cpu:0'):
    history = LSTM_model.fit(X_train, y_train, validation_split=0.25, epochs=10, batch_size=8)
```

```python
LSTM_model.summary()
```

```python
plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])
plt.title('model performance')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
prediction = LSTM_model.predict(X_test).flatten()

prediction = np.array(prediction)
```

```python
# visualisation
plt.figure(figsize = (30,10))
plt.plot(prediction, color = "b", label = "y_pred" )
plt.plot(y_test.flatten(), color = "g", label = "y_train")
plt.xlabel("Days")
plt.ylabel("Pressure")
plt.title("Simple RNN model, Predictions with input X_train vs y_train")
plt.legend()
plt.show()
```

```python
diff = np.abs(prediction - y_test.flatten())
diff
```

```python
T = 1.9
```

```python
faults = np.where(diff > T)[0]
faults = faults + TRAIN_SIZE
faults
```

```python
print("Total Faults in Data: ", len(np.arange(l_idx, h_idx+1)))
print("Predicted Faults: ",len(faults))
```

```python
correct_preds = np.intersect1d(faults, np.arange(l_idx, h_idx+1))
print("Correct predictions: ", len(correct_preds))
```

```python
TPR = (len(correct_preds) / len(faults))
```
