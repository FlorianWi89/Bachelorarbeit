```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

start = 7700
stop = 9500
step = 1
```

```python
s_id = 1

data = pd.read_parquet(f'TestData/SensorFaultScenariosTest_0/{s_id}_sensor_fault_test_data.parquet.gzip').astype(np.float64)
data = data['n54'].values

plt.figure(figsize=(15, 3))
plt.plot(data[start:stop:step])
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Pressure')
plt.savefig('Constant_sensor_fault.png', bbox_inches="tight")
plt.show()
```
